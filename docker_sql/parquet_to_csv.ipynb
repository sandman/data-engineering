```python
import pandas as pd
import os
```

```python
# Read the data
filename = 'yellow_tripdata_2021-01.parquet'
filepath = os.getcwd() + '/' + filename
df = pd.read_parquet(filepath)
```

```python
# Inspect the data
df.head()
```

```python
# Write to CSV
csv_output = filepath.replace('.parquet', '.csv')
df.to_csv(csv_output, index=False)
```

```python
# Write to CSV without compression
csv_filepath = filepath.replace('.parquet', '.csv')
df.to_csv(csv_filepath, index=False)
```

```python
# Write to CSV with compression
csv_filepath_compressed = filepath.replace('.parquet', '.csv.gz')
df.to_csv(csv_filepath_compressed, index=False, compression='gzip')
```

```python
df = pd.read_csv('yellow_tripdata_2021-01.csv', nrows=100)
print(pd.io.sql.get_schema(df, name="yellow_taxi_data"))
```
